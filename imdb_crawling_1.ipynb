```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/movies-coming-soon/2022-02/'
```

```python
res = requests.get(url)
print(res.text)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# title, genre, content, director, stars, rating...
```

```python
path = '#main > div > div.list.detail > div'
```

```python
tags = soup.select(path)
tags
```

```python
len(tags)
```

```python
title = tags[0].select('#main > div > div.list.detail > div:nth-child(2) > table > tbody > tr:nth-child(1) > td.overview-top > h4 > a')[0].text
```

```python
genres = []
for g in tags[0].select('#main > div > div.list.detail > div:nth-child(2) > table > tbody > tr:nth-child(1) > td.overview-top > p > span'):
    if g != '|':
        genres.append(g.text)
genres
```

```python
content = tags[0].select('#main > div > div.list.detail > div:nth-child(2) > table > tbody > tr:nth-child(1) > td.overview-top > div.outline')[0].text
```

```python
director = tags[0].select('#main > div > div.list.detail > div:nth-child(2) > table > tbody > tr:nth-child(1) > td.overview-top > div:nth-child(5) > span > a')[0].text
```

```python
stars = []

for s in tags[0].select('#main > div > div.list.detail > div:nth-child(2) > table > tbody > tr:nth-child(1) > td.overview-top > div:nth-child(6) > a'):
    stars.append(s.text)
stars
```

```python
time = tags[0].select('td.overview-top > p > time')[0].text
```

```python
movies = []

for tag in tags:
    
    title = tag.select('td.overview-top > h4 > a')[0].text
    
    
    genres = []
    for g in tag.select('td.overview-top > p > span'):
        if '|' not in g:
            genres.append(g.text)
    
    content = tag.select('td.overview-top > div.outline')[0].text    
    
    director = tag.select('td.overview-top > div:nth-child(5) > span > a')[0].text    
    
    stars = []

    for s in tag.select('td.overview-top > div:nth-child(6) > a'):
        stars.append(s.text)
        
        
    try:
        time = tag.select('td.overview-top > p > time')[0].text
    except:
        pass
    
    print('title :',title)
    print('genres :',genres)
    print('content :',content)
    print('director :',director)
    print('stars :',stars)
    if time:
        print('time :',time)
    
    print('-----------------')
    print()
    

    m = {}
    m['title'] = title
    m['genres'] = genres
    m['content'] = content
    m['director'] = director
    m['stars'] = stars
    m['time'] = time
    
    movies.append(m)
   
movies
```

```python
movies
```

```python
import pandas as pd
movies_df = pd.DataFrame(movies)
movies_df.to_csv('movies.csv')
```
