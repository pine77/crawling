```python
## crawling(scrapying)
# web에서 데이터를 프로그램을 통해서 가져오는 작업..

## web구성하는 언어 - html, css, javascript
# html은 웹을 기본 구조를 구성하는 스크립트 언어..
# css는 웹을 꾸며주는 언어
# javascript 웹이 동적으로 작동하도록 만드는 언어..

## html
# 기본 구조는 태그 - 여는 태그 '<', 닫는 태그 '/>'
# 태그 계층구조 부모태그, 자식태그..
# <head>  정보... </head> : 웹의 정보를 담고 있는 태그
# <body>  태그... </body> : 웹에서 눈으로 보는 내용이 들어있는 태그..

```

```python
<html>
    <head>
        web page meta정보    
    </head>
    
    <body>  # 웹에서 눈에 보이는 부분
        웹페이지에서 보여지는 콘텐츠가 들어가는 부분..
        <p>문단을 나타냅니다</p>    # p : 문단 태그
        <div><h3>제목을 나타냅니다</h3></div>   # div : 박스태그,  h : 제목태그
    </body>
</html>
```

```python
<html>

    <head>
        web page meta정보    
    </head>
    
    <body>
        <table border = 1>  #   table : 표를 만드는 태그
        <tr>    #  tr : 테이블의 행
            <td>항목</td>   #  td : 테이블의 칼럼
            <td>2013</td>
            <td>2014</td>
        </tr>
        <tr>
            <td>항목</td>
            <td>2013</td>
            <td>2014</td>
        </tr>
    </body>
    
</html>
```

```python
<html>
    <head>
        <style>   # css 스타일
            p {color:blue;}
            div {color : red;}
        </style>        
    </head>
    
    <body>
        <p> paragragh1 </p>
        <p> paragragh2 </p>
        <div> div1 </div>        
    </body>
</html>
```

```python
<html>
    <head>
        <style>
            #title{color:blue;}   # id : 전체 웹문서에서 단 한 곳에만 적용 가능
            .par{color : red;}    # class : 여러 곳에 적용이 가능
        </style>        
    </head>
    
    <body>
        <p id=title> paragragh1 </p>
        <p class = par> paragragh2 </p>
        <div class=par> div1 </div>        
    </body>
</html>
```

```python
# 검색키워드를 입력받은 후 제목과 url,  본문을 크롤링 하는 함수를 만드시오..

def crawling_daum_news(kw):
    url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q={}'
    url = url.format(kw)
    print(url)
    
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    
    path = '#newsColl > div.cont_divider > ul > li'
    tags = soup.select(path)
    
    for tag in tags:
        print(tag.select('a.tit_main')[0].text)
        print(tag.select('a.tit_main')[0]['href'])
        print(tag.select('p.desc'))
        print('---------')
        print()
```

```python
def crawling_daum_news(kw, p):
    
    for i in range(1, p+1):
        url = 'https://search.daum.net/search?nil_suggest=btn&w=news&DA=PGD&q={}&p={}'
        url = url.format(i, kw)
        print(url)
        
        html = requests.get(url)
        soup = BeautifulSoup(html.text, 'html.parser')

        path = '#newsColl > div.cont_divider > ul > li'
        tags = soup.select(path)
#         print(tags)

        for tag in tags:
            print(tag.select('a.tit_main')[0].text)
            print(tag.select('a.tit_main')[0]['href'])
            print(tag.select('p.desc'))
            print('Press:',tag.select('span.f_nb')[0].text)
            print('Date:',tag.select('span.f_nb')[1].text)
            print('---------')
            print()
```

```python
kw = '인공지능'
n = 10

crawling_daum_news(kw, n)
```
